```python
from nltk.corpus import stopwords
```

```python
stopwords_trump = inaugural.words(fileids='2017-Trump.txt')
type(stopwords_trump)
```

```python
trump_fre = nltk.FreqDist(stopwords_trump)
```

```python
for word in trump_fre:
    print(word, trump_fre[word])
```

```python
from nltk.probability import ConditionalFreqDist
cfd = ConditionalFreqDist((len(word), word) for word in inaugural.words(fileids='2017-Trump.txt')) #no of times a word is repeated in the corpus
```

```python
cfd[13] # words with length 13
```
